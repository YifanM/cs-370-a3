```python
import numpy as np
import matplotlib.pyplot as plt

from numpy.fft import fft2, ifft2
```

```python
f = np.array(plt.imread("operahall.png"), dtype=float)
plt.imshow(f, cmap="gray");
```

```python
top_left_block = f[:32, :32]
plt.imshow(top_left_block, cmap="gray");
```

```python
F_top_left = fft2(top_left_block)
plt.imshow(abs(F_top_left), cmap="gray")
```

The brightest pixel is by far the one in the very top left corner at position $(0, 0)$. This pixel is special in that it represents the average pixel intensity in the original block. Since the remaining magnitudes in the Fourier transform are all black, we can say that the average pixel intensity in the original block is much larger than the magnitudes of the frequencies in the Fourier transform.

```python
F_top_left[0][0] = 0
plt.imshow(abs(F_top_left), cmap="gray")
```

```python
def Compress(X, tol):
    N = len(X)
    total = 0
    dropped = 0
    Y = np.zeros((N, N))
    
    for i in range(0, N, 32):
        for j in range(0, N, 32):
            # Get the block and FFT it
            block = X[i:i+32,j:j+32]
            F_block = fft2(block)
            
            # Update original number of non-zero coefficients
            total += np.count_nonzero(abs(F_block))
            
            # Find max magnitude
            F_max = abs(F_block).max()
            
            # If magnitude is lower than max * tol, set to 0
            for k in range(32):
                for l in range(32):
                    if abs(F_block[k][l]) <= F_max * tol and abs(F_block[k][l]) != 0:
                        F_block[k][l] = 0
                        dropped += 1
            
            # Restore the compressed block
            compressed_block = np.real(ifft2(F_block))
            Y[i:i+32,j:j+32] = compressed_block
    
    return (Y, dropped / total)
```

```python
# Drop rate of around 0.5

compressed = Compress(f, 0.001)
print("Drop rate: " + "{0:0.5f}".format(compressed[1]))
plt.imshow(compressed[0], cmap="gray");
```

```python
# Drop rate of around 0.8

compressed = Compress(f, 0.0045)
print("Drop rate: " + "{0:0.5f}".format(compressed[1]))
plt.imshow(compressed[0], cmap="gray");
```

```python
# Drop rate of around 0.92

compressed = Compress(f, 0.0095)
print("Drop rate: " + "{0:0.5f}".format(compressed[1]))
plt.imshow(compressed[0], cmap="gray");
```

```python
# Drop rate of around 0.97

compressed = Compress(f, 0.02)
print("Drop rate: " + "{0:0.5f}".format(compressed[1]))
plt.imshow(compressed[0], cmap="gray");
```

```python
# Error between the original and the highest quality compression (ie drop rate = 0.5)

print("Error when drop rate is 0.5")
compressed = Compress(f, 0.001)
error = abs(f - compressed[0])
plt.imshow(error, cmap="gray");
plt.show()

# Error between the original and the lowest quality compression (ie drop rate = 0.97)

print("Error when drop rate is 0.97")
compressed = Compress(f, 0.02)
error = abs(f - compressed[0])
plt.imshow(error, cmap="gray");
plt.show()
```

We can see the error when the drop rate is only 0.5 is fairly insignificant. When the drop rate is 0.97 however, we can see an outline of the original picture. This indicates that the most loss occurred at the borders between objects in the original picture.
